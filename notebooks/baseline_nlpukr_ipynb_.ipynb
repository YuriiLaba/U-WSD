```python
nlp = stanza.Pipeline(lang='uk', processors='tokenize,mwt,pos', verbose = False)
text = "Олені скупчилися на відстої"
doc = nlp(text)
print(*[f'word: {word.text}\tupos: {word.upos}\txpos: {word.xpos}\tfeats: {word.feats if word.feats else "_"}' for sent in doc.sentences for word in sent.words], sep='\n')
```

```python
def find_target_word_in_sentence(word, text):
  tokens = udpipe_model.tokenize(text)

  for tok_sent in tokens:
    udpipe_model.tag(tok_sent)

    for word_index, w in enumerate(tok_sent.words[1:]):
      if w.lemma == word:
        print(w.upostag)
        return word_index
  return None

def get_pos_by_context(word, text):
  doc = nlp(text)
  word_index = find_target_word_in_sentence(word, text)
  if word_index is not None:
    return doc.sentences[0].words[word_index].upos
  else:
    return None
```
